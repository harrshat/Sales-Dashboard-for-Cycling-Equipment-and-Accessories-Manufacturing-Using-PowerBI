```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Using Iris Dataset
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
#data:target
# 4 independent features 'data' and one output feature -'target'
```

```python
iris['target']
```

```python
print(iris['DESCR'])

'''
class:
            - Iris-Setosa
            - Iris-Versicolour
            - Iris-Virginica
'''
```

```python
#Independent features
X=pd.DataFrame(iris['data'],columns=['sepal_length','sepal_width','petal_length','petal_width'])
#dependent features
y=pd.DataFrame(iris['target'])
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=42)
```

```python
#Apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
treeClassifier=DecisionTreeClassifier()
'''
criterion: str = "gini",
    splitter: str = "best",
    max_depth: Any | None = None,
    min_samples_split: int = 2,
    min_samples_leaf: int = 1,
    min_weight_fraction_leaf: float = 0,
    max_features: Any | None = None,
    random_state: Any | None = None,
    max_leaf_nodes: Any | None = None,
    min_impurity_decrease: float = 0,
    class_weight: Any | None = None,etc.
  '''
```

```python
treemodel=treeClassifier.fit(X_train,y_train)
treemodel
```

```python
#Visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
#x[3]--3rd feature
```

```python
# To avoid overfitting, we do post pruning/ pre pruning
# Post pruning - Construct the decision tree at max depth
# Pre pruning - hyper parameter tuning such as max_depth, etc.
```

```python
treeClassifier2=DecisionTreeClassifier(max_depth=2)
treemodel2=treeClassifier2.fit(X_train,y_train)
treemodel2
```

```python
tree.plot_tree(treeClassifier2)
```

```python
y_pred=treeClassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm= confusion_matrix(y_test,y_pred)
print("Confusion Matrix:\n",cm)
print("\n")
print("__________________________________________________________________")
cr=classification_report(y_test,y_pred)
print("Classification report:\n",cr)
```
